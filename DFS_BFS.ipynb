AIM: WAP to implement DFS, BFS procedures to search a node in a graph

Depth First Search (DFS) Algorithm

Step 1: Set count = 0

Step 2: Create a stack and push the starting node s

Step 3: Mark s as visited

Step 4: While stack is not empty do
     
      4.1 Pop the top node n from stack

      4.2 Increase count = count + 1

      4.3 If n = x then return True and count

      4.4 For each neighbor m of n
      - If m is not visited then mark m visited and push m into stack

Step 5: End while

Step 6: Return False and count

#Breadth First Search (BFS) Algorithm

Step 1: Set count = 0

Step 2: Create a queue and enqueue the starting node s

Step 3: Mark s as visited

Step 4: While queue is not empty do
      4.1 Dequeue the front node n

      4.2 Increase count = count + 1

      4.3 If n = x then return True and count

      4.4 For each neighbor m of n

      - If m is not visited then mark m visited and enqueue m

Step 5: End while

Step 6: Return False and count

```python
#Coding

from collections import deque

def dfs(graph, start, target):
    visited = set()
    stack = [start]
    count = 0

    while stack:
        node = stack.pop()
        count += 1
        if node == target:
            return True, count
        if node not in visited:
            visited.add(node)
            stack.extend(graph.get(node, []))
    return False, count


def bfs(graph, start, target):
    visited = set()
    queue = deque([start])
    count = 0

    while queue:
        node = queue.popleft()
        count += 1
        if node == target:
            return True, count
        if node not in visited:
            visited.add(node)
            queue.extend(graph.get(node, []))
    return False, count


# --- User Input ---
n = int(input("Enter number of edges: "))
graph = {}
for _ in range(n):
    u, v = input("Enter edge (u v): ").split()
    graph.setdefault(u, []).append(v)
    graph.setdefault(v, []).append(u)  # undirected graph

start = input("Enter starting node: ")
target = input("Enter target node: ")

found_dfs, count_dfs = dfs(graph, start, target)
found_bfs, count_bfs = bfs(graph, start, target)

print("\nDFS Search Result:", found_dfs, "Comparisons:", count_dfs)
print("BFS Search Result:", found_bfs, "Comparisons:", count_bfs)
```
