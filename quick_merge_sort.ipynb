AIM: WAP for the Quick Sort and Merge Sort using Count Variable

Quick Sort Algorithm (with Count)
Algorithm Steps:

Step 1: Choose a pivot (commonly the last element).

Step 2: Partition the array:
        
        -Compare each element with the pivot.
        -Increment count for each comparison.
        -Place smaller elements on the left, larger on the right.

Step 3: Recursively apply Quick Sort to left and right subarrays.

Step 4: Stop when subarray size â‰¤ 1.

Step 5: Return sorted array and total comparisons.

Merge Sort Algorithm (with Count)
Algorithm Steps:

Step 1:Divide the array into two halves.

Step 2: Recursively sort each half.

Step 3: Merge the two halves:

      - Compare elements from left and right arrays.
      - Increment count for each comparison.
      - Place the smaller element into the merged array.

Step 4: Continue until all elements are merged.

Step 5: Return sorted array and total comparisons.

#Quick Sort Code

```python
def quick_sort(arr):
    count = [0]  # mutable counter

    def partition(low, high):
        pivot = arr[high]
        i = low - 1
        for j in range(low, high):
            count[0] += 1  # comparison
            if arr[j] <= pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
        arr[i+1], arr[high] = arr[high], arr[i+1]
        return i+1

    def quick(low, high):
        if low < high:
            pi = partition(low, high)
            quick(low, pi-1)
            quick(pi+1, high)

    quick(0, len(arr)-1)
    return arr, count[0]


# --- User Input ---
arr = list(map(int, input("Enter numbers separated by space: ").split()))
sorted_arr, comparisons = quick_sort(arr)
print("\nQuick Sort Result:", sorted_arr)
print("Comparisons:", comparisons)
```

#Merge Sort Code

```python
def merge_sort(arr):
    count = [0]

    def merge(left, right):
        merged = []
        i = j = 0
        while i < len(left) and j < len(right):
            count[0] += 1  # comparison
            if left[i] <= right[j]:
                merged.append(left[i])
                i += 1
            else:
                merged.append(right[j])
                j += 1
        merged.extend(left[i:])
        merged.extend(right[j:])
        return merged

    def sort(sub_arr):
        if len(sub_arr) <= 1:
            return sub_arr
        mid = len(sub_arr) // 2
        left = sort(sub_arr[:mid])
        right = sort(sub_arr[mid:])
        return merge(left, right)

    return sort(arr), count[0]


# --- User Input ---
arr = list(map(int, input("Enter numbers separated by space: ").split()))
sorted_arr, comparisons = merge_sort(arr)
print("\nMerge Sort Result:", sorted_arr)
print("Comparisons:", comparisons)
```
