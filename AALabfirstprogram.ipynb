#Insertion Sort
Idea: Build the sorted array one element at a time by inserting each new element into its correct position among the already sorted elements.

Steps:

Step 1 Assume the first element is sorted.

Step 2 Take the next element and compare it with elements in the sorted portion.

Step 3 Shift larger elements one position to the right.

Step 4 Insert the current element into its correct position.

Step 5 Repeat until all elements are sorted.

#Selection sort

Idea: Repeatedly find the smallest (or largest) element from the unsorted part of the array and put it at the beginning.

Steps:

Step 1 Start from the first element.

Step 2 Find the minimum element in the unsorted portion.

Step 3 Swap it with the first element of the unsorted portion.

Step 4 Move the boundary of the sorted portion one step forward.

Step 5 Repeat until the array is sorted.

```python
def selection_sort(arr):
    """
    Selection Sort Algorithm
    Always does n(n-1)/2 comparisons regardless of input order.
    """
    n = len(arr)
    count = 0
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            count += 1   # counting comparison
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]  # swap
    return arr, count


def insertion_sort(arr):
    """
    Insertion Sort Algorithm
    Comparisons/shifts depend on input order (best, average, worst).
    """
    n = len(arr)
    count = 0
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            count += 1   # counting comparison/shift
            arr[j+1] = arr[j]
            j -= 1
        if j >= 0:       # one extra comparison when while loop fails
            count += 1
        arr[j+1] = key
    return arr, count


# Main Program
arr = list(map(int, input("Enter numbers separated by space: ").split()))

# Best Case: already sorted
best_case = sorted(arr)
sel_best, sel_count_best = selection_sort(best_case.copy())
ins_best, ins_count_best = insertion_sort(best_case.copy())

# Average Case: original input order
avg_case = arr.copy()
sel_avg, sel_count_avg = selection_sort(avg_case.copy())
ins_avg, ins_count_avg = insertion_sort(avg_case.copy())

# Worst Case: reverse sorted
worst_case = sorted(arr, reverse=True)
sel_worst, sel_count_worst = selection_sort(worst_case.copy())
ins_worst, ins_count_worst = insertion_sort(worst_case.copy())

# Output
print("Selection Sort")
print("Best Case (Sorted):", sel_best, "Comparisons:", sel_count_best)
print("Average Case (Input):", sel_avg, "Comparisons:", sel_count_avg)
print("Worst Case (Reverse):", sel_worst, "Comparisons:", sel_count_worst)

print("Insertion Sort")
print("Best Case (Sorted):", ins_best, "Comparisons/Shifts:", ins_count_best)
print("Average Case (Input):", ins_avg, "Comparisons/Shifts:", ins_count_avg)
print("Worst Case (Reverse):", ins_worst, "Comparisons/Shifts:", ins_count_worst)
```
