#Program No. 03
AIM: WAP to implement Binary Search using count variable

Step 1: Set low = 0, high = n - 1, count = 0

Step 2: While low â‰¤ high do
      
      - Compute mid = (low + high) // 2
      
      
      - Increase count = count + 1

      - If arr[mid] = x then return mid and count

      - Else if arr[mid] < x then set low = mid + 1

      - Else set high = mid - 1

Step 3: End while

Step 4: Return -1 and count

```python
def binary_search(arr, x):
    low, high = 0, len(arr) - 1
    count = 0

    while low <= high:
        mid = (low + high) // 2
        count += 1

        if arr[mid] == x:
            return mid, count
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1

    return -1, count


#User Input
arr = list(map(int, input("Enter sorted numbers: ").split()))
x = int(input("Enter number to search: "))

index, comparisons = binary_search(arr, x)

if index != -1:
    print("Element found at index", index)
else:
    print("Element not found")
print("Comparisons:", comparisons)
```
